# *Pandas*

Pandas is the most popular python library that is used for data analysis. It provides highly optimized performance with back-end source code is purely written in C or Python. We can analyze data in pandas with: Series.

```python
import pandas as pd
```

```python
#how to create Series?
l=[1,2,3,4,5,6]
pd.Series(l,index=['a','b','c','d','e','f'])
```

```python
t=(1,2,3)
s=pd.Series(t,index=['a','b','c'])
s
```

```python
#how to create Dataframe
```

```python
# creating dataframe with one column
l=[1,2,3,4]
pd.DataFrame(l,columns=['sno'])
```

```python
#creating dataframe with multiple columns
df=pd.DataFrame({'name':['emmanuel','simon','hero'],'class':[1,2,3],'marks':[22,23,44]})
df.head()
```

```python
#how to add new columns to dataframe?
df['lang']=pd.DataFrame(['english','hindi','kannada'])
df.head()
```

```python
#how to change values of dataframe
df['marks']=pd.DataFrame([44,55,66])
df.head()
```

```python
df['class']=pd.DataFrame([9,10,8])
df.head()
```

```python
#how to sort values of column in dataframe
df.sort_values(by='class')
```

```python
df.sort_values(by='marks')
```

```python
#how to use groupby on column
df.groupby('lang').sum()
```

```python
df.groupby('name').sum()
```

```python
# # groupby can also be used like this 
# df.groupby('name').mean()
# df.groupby('name').median()
# df.groupby('name').mode()
```

```python
#groupby with multiple columns
df.groupby(['name','marks']).sum()
# can also be used .mean(),.median(),.mode()
```

```python
# Converting the raw grades to a categorical data type:
df.marks.astype('float')
```

```python
df.marks.astype('int')
```

```python
df.marks.astype('category')
```

```python
# merging two columns
```

```python
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})
left
```

```python
right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})
right
```

```python
pd.merge(left,right,on='key')
```

```python
df1=pd.DataFrame({'name':['emmanuel','simon','superman','krish'],'rollno':[1,2,3,4],'marks':[10,20,0,40]})
df1
```

```python
df2=pd.DataFrame({'name':['emmanuel','simon','superman','krish'],'rollno':[1,2,3,4],'marks':[100,200,00,400]})
df2
```

```python
pd.merge(df1,df2,on='rollno')
```
